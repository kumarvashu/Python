```python
import pandas as pd
from IPython.display import display

excel_file_location = "/Users/yashkadam/Downloads/Badly_Structured_Sales_Data_1.xlsx"

raw_df = pd.read_excel(excel_file_location, header=[0, 1, 2], index_col=0)

# Stack the columns to create a multi-indexed DataFrame
stacked_df = raw_df.stack(level=[0, 1, 2]).reset_index()

# Rename the headers (column names)
new_column_names = {
    'level_0': 'Order ID',
    'Segment>>': 'Segment',
    'Ship Mode>>':'Ship Mode',
    'Order ID': 'To be deleted',
    0: 'Sales'
}

stacked_df = stacked_df.rename(columns=new_column_names)

stacked_df
```

```python
# Delete the 'To be deleted' column
stacked_df = stacked_df.drop('To be deleted', axis=1)
stacked_df
```

```python
# Remove the rows containing total and Grand total
stacked_df = stacked_df[~stacked_df['Segment'].str.contains('Home Office Total|Consumer Total|Corporate Total')]
stacked_df = stacked_df[~stacked_df['Order ID'].str.contains('Grand Total')]
# sorting
stacked_df=stacked_df.sort_values(by=['Segment','Ship Mode'])
stacked_df
```
